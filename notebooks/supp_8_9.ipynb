## S8: Fig. 1 with only PPEs

```python
xvar = 'T_S'
yvar_1 = 'ET'
yvar_2 = 'P'

# Create figure and axes
fig, axes = plt.subplots(
    3, 2, figsize=(8, 7),
    width_ratios=[1,1],
    height_ratios=[1,1,0.08],
    layout='tight',
)
ax = axes.flatten()

## Create separate axis for the legend
gs = ax[4].get_gridspec()
for a in ax[4:]:
    a.remove()
axleg = fig.add_subplot(gs[4:])

## Set up axis limits
xlim = [
    [-3.1, 1.50],
    [-1.0, 0.75],
    [-3.1, 1.50],
    [-1.0, 0.75],
]
ylim = [
    [-0.20, 0.75],
    [-0.20, 0.75],
    [-0.20, 0.75],
    [-0.20, 0.75],
]

these_stats = dict(hadhst=dict(), hada1b=dict(), cesm=dict(), cmip=dict())

for i, sm in enumerate(['lm', 'om']):
    ##
    key = "cesm"
    x = cesm_delta_areamean[sm][xvar]
    
    y = cesm_delta_areamean[sm][yvar_1]
    these_stats[key][sm+"_"+yvar_1] = myplt.plot_scatter_regression(ax[i], x, y, myplt.odr_single, key, do_print=True)

    y = cesm_delta_areamean[sm][yvar_2]
    these_stats[key][sm+"_"+yvar_2] = myplt.plot_scatter_regression(ax[i+2], x, y, myplt.odr_single, key, do_print=True)

    ##
    key = "hadhst"
    x = hadhst_delta_areamean[sm][xvar]
    
    y = hadhst_delta_areamean[sm][yvar_1]
    these_stats[key][sm+"_"+yvar_1] = myplt.plot_scatter_regression(ax[i], x, y, myplt.odr_single, key, do_print=True)

    y = hadhst_delta_areamean[sm][yvar_2]
    these_stats[key][sm+"_"+yvar_2] = myplt.plot_scatter_regression(ax[i+2], x, y, myplt.odr_single, key, do_print=True)

    ##
    key = "hada1b"
    x = hada1b_delta_areamean[sm][xvar]
    
    y = hada1b_delta_areamean[sm][yvar_1]
    these_stats[key][sm+"_"+yvar_1] = myplt.plot_scatter_regression(ax[i], x, y, myplt.odr_single, key, do_print=True)

    y = hada1b_delta_areamean[sm][yvar_2]
    these_stats[key][sm+"_"+yvar_2] = myplt.plot_scatter_regression(ax[i+2], x, y, myplt.odr_single, key, do_print=True)

## Assign axis limits and tick labels
ax[3].set_ylim(None, 0.07)

## Add axis labels
for a in ax[:2]:
    a.set_ylabel('$\delta E$ [mm day$^{-1}$]', fontsize=10)
    a.set_xlabel('$\delta T_S$ [K]', fontsize=10)

for a in ax[2:]:
    a.set_ylabel('$\delta P$ [mm day$^{-1}$]', fontsize=10)
    a.set_xlabel('$\delta T_S$ [K]', fontsize=10)

ax[0].text(0.5, 1.1, 'Land mean', fontsize=12, fontweight='bold', ha="center", va="bottom", transform=ax[0].transAxes)
ax[1].text(0.5, 1.1, 'Ocean mean', fontsize=12, fontweight='bold', ha="center", va="bottom", transform=ax[1].transAxes)

## Add description of fit statistics to each subplot
def format_label_text(this_stats, key, alpha=0.05):
    ast = ""
    if this_stats["slope_p_value"] < alpha:
        ast = "^*"
    this_label = f"{myplt.label_dict[key]}: ${this_stats['slope']: 0.3f} \\pm {this_stats['slope_ci_halfwidth']: 0.3f}{ast}$"
    return this_label

stats_xpos = [0.55, 0.04, 0.55, 0.55]
stats_ypos = 0.775
stats_width = 0.46
stats_height = 0.25

for i, v, sm in zip(range(4), ["ET", "ET", "P", "P"], ["lm", "om", "lm", "om"]):
    for j, key in enumerate(["cesm", "hadhst", "hada1b"]):
        ax[i].text(
            stats_xpos[i], stats_ypos+(3-j)*(stats_height/4),
            format_label_text(these_stats[key][sm+"_"+v], key, alpha=0.05),
            fontsize=6, color=myplt.c_dict[key],
            ha='left', va='top',
            transform=ax[i].transAxes
        )

## Add legend
h, l = ax[0].get_legend_handles_labels()
leg = axleg.legend(h, l, ncol=4, fontsize=9, loc='center', markerscale=2)
for line in leg.get_lines():
    line.set_linewidth(3.0)
axleg.axis('off')

## Add subplot numbering
for i,a in enumerate(ax[:4]):
    a.set_title(f'({myplt.figenum[i]})', loc="left", fontsize=12, fontweight='regular')

## Save figure
plt.show()
# fig.savefig(f'supp_delta_e_delta_p.png', dpi=export_dpi, bbox_inches='tight')
fig.savefig(f'delta_e_delta_p_ppe_zoomedin.pdf', bbox_inches='tight')
# fig.savefig(f'supp_delta_e_delta_p.svg', bbox_inches='tight')
```
